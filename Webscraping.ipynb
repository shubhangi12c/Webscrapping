```python
import requests
```

```python
url_link = "https://en.wikipedia.org/wiki/Data_analysis"
```

```python
response = requests.get(url=url_link)
response.status_code
#here we are checking that website is allowing for scrapping or not
```

```python
from bs4 import BeautifulSoup
```

```python
data = response.content
print(data)
```

```python
soup = BeautifulSoup(data,"html.parser")
```

```python
soup.find("title")
```

```python
title = soup.find("title").text
title
```

```python
headers = soup.find_all("h1",class_="firstHeading mw-first-heading")
```

```python
headers
```

```python
ptags = soup.find_all("p")
for p in ptags:
    print(p)
```

```python
ptags[2]
```

```python
ptags[0].text
```

```python
#p_content =[p.text for p in ptags]
p_content = []
for p in ptags:
    p_content.append(p.text)
print(p_content)
```

```python
other_headers = soup.find_all("div",class_="mw-heading mw-heading2")
oth_headers = [header.text for header in other_headers]
print(oth_headers)
```

```python
img_tags = soup.find_all("a",class_="mw-file-description")
img_tags
```

```python
img_tags[0].get("href")
```

```python
base_link = "https://en.wikipedia.org/wiki/Data_analysis/"
img_links = []
for links in img_tags:
    link = links.get("href")
    img_links.append(base_link+link)
```
